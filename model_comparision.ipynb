```python
import json
import pandas as pd

# Load saved results
with open("../results/model_comparison.json", "r") as f:
    results = json.load(f)

# Convert dict to DataFrame
comparison_df = pd.DataFrame(results).T

print("Loaded model comparison results.")
comparison_df
```

```python
comparison_ranked = comparison_df.sort_values(by="roc_auc", ascending=False)
comparison_ranked
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(comparison_df.index, comparison_df["roc_auc"])
plt.title("Model Comparison (ROC-AUC)")
plt.ylabel("ROC-AUC")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.bar(comparison_df.index, comparison_df["accuracy"], color="orange")
plt.title("Model Comparison (Accuracy)")
plt.ylabel("Accuracy")
plt.xticks(rotation=45, ha="right")
plt.show()
```

```python
best_model = comparison_ranked.index[0]
print("Best Model:", best_model)
```
